```sql
USE CATALOG my_catalog;
USE SCHEMA school;

SELECT current_catalog(),current_schema()
```

```sql
drop table if exists student_temp;
drop table if exists student_temp_BROZE;
CREATE TABLE student_temp_BROZE (
  id INT,
  name STRING,
  country STRING,
  state STRING,
  latitude DOUBLE,
  longitude DOUBLE,
  gender STRING,
  ethnic_group STRING,
  age INT,
  english_score DOUBLE,
  math_score DOUBLE,
  sciences_score DOUBLE,
  language_score DOUBLE
)
USING DELTA;
```

```sql
COPY INTO student_temp_BROZE
FROM '/Volumes/my_catalog/school/student_raw_data/student-dataset.csv'
FILEFORMAT = CSV
FORMAT_OPTIONS(
  'header' = 'true',
  'inferSchema' = 'true'
)
COPY_OPTIONS('mergeSchema' = 'true');
```

```sql
-- BRONZE DONE
select * from student_temp_BROZE;
```

```sql
CREATE OR REPLACE TABLE student_temp_SILVER AS  SELECT ID,NAME,round(latitude,1) as latitude,
round(longitude,1) as longitude,decode(gender,'F','Female','M','Male') as gender,AGE ,nationality
,city,english_grade,math_grade,sciences_grade,language_grade FROM student_temp_BROZE;
```

```sql
SELECT * FROM student_temp_SILVER;
```

```sql
create or replace temp view  student_country_cnt as select nationality,count(1) from student_temp_SILVER from group by nationality
```

```sql
select * from student_country_cnt;
```

```sql
create table if not exists student_temp_gold (STUDENT_COUNTRY string,No_of_students long);
```

```sql
insert overwrite table student_temp_gold select * from student_country_cnt;
```

```sql
select * from student_temp_gold order by No_of_students desc;
```
